```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot(data, label):
    MARKERSIZE = 2
    plt.figure(figsize = (10, 10))
    plt.plot(data['train_ground_truth'], data['train_predictions'], 'bo', markersize = MARKERSIZE, alpha = 0.1, label = 'train')
    plt.plot(data['val_ground_truth'], data['val_predictions'], 'ro', markersize = MARKERSIZE, alpha = 0.1, label = 'val')
    if 'train_manifold_ground_truth' in data.keys():
        plt.plot(data['train_manifold_ground_truth'], data['train_manifold_predictions'], 'o', color = 'black', 
                 markersize = MARKERSIZE, label = 'train_manifold')
    if 'val_manifold_ground_truth' in data.keys():
        plt.plot(data['val_manifold_ground_truth'], data['val_manifold_predictions'], 'o', color = 'green', 
                 markersize = MARKERSIZE, label = 'val_manifold')
    plt.legend()
    plt.title(label)
    plt.xlabel("ground truth")
    plt.ylabel("predictions")
    plt.show()
```

```python
data_50 = np.load('using_20_from_manifold.npz')
plot(data_50, 'using 20 from manifold')
```

```python
data_full = np.load('use_full_manifold.npz')
plot(data_full, 'using full manifold')
```

```python
data_not = np.load('not_using_manifold.npz')
plot(data_not, 'not using manifold')
```
